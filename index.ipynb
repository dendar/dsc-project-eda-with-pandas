```python
# Let's get started importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
# Loading the data
df = pd.read_csv("ames_train.csv")
```

```python
# Investigate the Data
df.info() "LotArea"
```

```python
# Investigating Distributions using scatter_matrix
pd.plotting.scatter_matrix(df[["YearBuilt","FullBath","SalePrice","YrSold"]])
```

```python
# Create a plot that shows the SalesPrice Distribution
fix = plt.subplots(figsize = (10,8))
plt.hist(df["SalePrice"])
plt.title("Distribution of sale")
plt.xlabel("sales price")
plt.ylabel("Number of House")
plt.axvline(df["SalePrice"].mean(), color="g")
plt.axvline(df["SalePrice"].median(), color="y")
plt.axvline(df["SalePrice"].std(), color="r")
```

```python
# Create a plot that shows the LotArea Distribution
fig = plt.subplots(figsize = (8,6))

plt.hist(df["LotArea"])
plt.title("LotArea of Distribution")
plt.xlabel('LotArea')
plt.ylabel("Number of house")
plt.axvline(df["LotArea"].mean(), color="g")
df.LotArea.mean()
```

```python
# Create a plot that shows the Distribution of the overall house condition
x= df.OverallCond.unique()
y = df["OverallCond"].value_counts()
fig= plt.subplots(figsize=(6,6))

plt.title("Distribution of the overall house condition")
plt.xlabel("condition of house")
plt.ylabel("number of house")
plt.hist(df.OverallCond, bins='auto')
plt.show()
```

```python
# Create a Box Plot for SalePrice
plt.boxplot(df.SalePrice)
plt.show()
```

```python
df.corr()['SalePrice'].sort_values()
```

```python
# Perform an Exploration of home values by age
df["age"] = df["YrSold"]- df["YearBuilt"]
df["decades"] = df.age//10
to_plot = df.groupby('decades').SalePrice.mean()
to_plot.plot(kind="bar", figsize =(10, 10))
plt.ylabel("House age in Decades")
plt.xlabel("Average Sale Price of Homes")
plt.title("Average Home Values by Home Age")
plt.show()
```
